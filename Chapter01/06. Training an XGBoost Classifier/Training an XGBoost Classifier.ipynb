# XGBoost(Extreme Gradient Boosting) 분류기
- 경사부스팅(Gradient Boosting)은 일반적인 머신러닝 문제에 대하여 가장 신뢰할 수 있으며 정확한 알고리즘으로 알려져 있다.
- XGBoost를 사용해 악성코드 탐지기를 만들 수 있다.

```python
import pandas as pd

df = pd.read_csv("file_pe_headers.csv", sep=",")
y = df["Malware"]
X = df.drop(["Name", "Malware"], axis=1).to_numpy()
```

```python
# 데이터셋을 훈련 데이터와 테스트 데이터로 분할
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# XGBoost 모델의 인스턴스를 하나 만들어 훈련 데이터셋으로 훈련
from xgboost import XGBClassifier

XGB_model_instance = XGBClassifier()
XGB_model_instance.fit(X_train, y_train)
```

```python
# 훈련 모델을 이용하여 테스트 데이터셋을 이용해 예측
y_test_pred = XGB_model_instance.predict(X_test)
```

```python
# 테스트 데이터셋 모델의 성능을 평가(정확도 계산)
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_test_pred)
print("Accuracy: %.4f%%" % (accuracy * 100))
```
